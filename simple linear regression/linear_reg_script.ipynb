```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python

df = pd.read_csv("honeyproduction.csv")
print(df.head())
```

```python
prod_per_year = df.groupby('year')['totalprod'].mean().reset_index()
```

```python
X = prod_per_year['year'].values.reshape(-1, 1)  
y = prod_per_year['totalprod']  
```

```python
# Step 4: Scatterplot of actual data
plt.scatter(X, y, label='Actual Data', color='blue')
plt.xlabel('Year')
plt.ylabel('Total Production')
plt.title('Total Honey Production per Year')
plt.legend()
plt.show()
```

```python
regr = LinearRegression()
regr.fit(X, y)

print(f"Slope (coefficient): {regr.coef_[0]:.2f}")
print(f"Intercept: {regr.intercept_:.2f}")
```

```python
y_predict = regr.predict(X)
```

```python
plt.scatter(X, y, label='Actual Data', color='blue')
plt.plot(X, y_predict, color='red', label='Regression Line')
plt.xlabel('Year')
plt.ylabel('Total Production')
plt.title('Honey Production Trend')
plt.legend()
plt.show()
```

```python
X_future = np.arange(2013, 2051).reshape(-1, 1)
future_predict = regr.predict(X_future)
```

```python
plt.plot(X_future, future_predict, color='green', label='Future Prediction')
plt.xlabel('Year')
plt.ylabel('Total Production')
plt.title('Future Honey Production Predictions')
plt.legend()
plt.show()
```

The downward trend (negative slope) suggests that honey production is predicted to decrease consistently year after year.
